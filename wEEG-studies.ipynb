# Wireless EEG systems
Plot histograms of the data

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Major ticks and minor ticks
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
# Open Googlesheet csv file
df = pd.read_csv('/Users/guiomar/Downloads/Data_Figures - Studies-2.csv')
df;
df.head(1)
```

## Number of channels
Number of channels used in the reviewed studies

```python
FS = 22
plt.style.use('seaborn')
```

```python
fig = plt.figure()
ax  = plt.axes()
plt.hist(df['Nchannels'], bins=32, density=False, label='Nchannels')

# Add title
plt.title('Number of channels', fontweight='bold',fontsize=FS)

# Add data labels
ax.set_xlabel('Number of channels',fontsize=FS-2)
ax.set_ylabel('Number of studies',fontsize=FS-2)

# Set the ranges to plot of x
ax.set_xlim([0, 70])
ax.set_ylim([0, 35])


ax.tick_params(axis='both', which='major', labelsize=16)
ax.xaxis.set_minor_locator(MultipleLocator(2))
ax.tick_params(which='minor', length=3, color='k', direction='out')
ax.tick_params(which='major', length=5, color='k')



plt.show();
```

## Year of the studies
Year of the reviewed studies

```python
fig = plt.figure()
ax  = plt.axes()
plt.hist(df['Year'], bins=25, density=False, label='Year')
#sns.distplot(df['Year'], bins=20)

# Add title
plt.title('Number of studies using wEEG', fontweight='bold',fontsize=FS)
#plt.style.use('seaborn')

# Add data labels
ax.set_xlabel('Year of the study',fontsize=FS-2)
ax.set_ylabel('Number of studies',fontsize=FS-2)


# Set the ranges to plot of x
ax.set_xlim([1995, 2020])
ax.set_ylim([0, 22.5])


ax.tick_params(axis='both', which='major', labelsize=16)
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.tick_params(which='minor', length=3, color='k', direction='out')
ax.tick_params(which='major', length=5, color='k')

plt.show()
```

## Number of participants
Number of participants used in the reviewed studies

```python
cond = df['Nparticipants']<500
```

```python
fig = plt.figure()
ax  = plt.axes()
plt.hist(df['Nparticipants'][cond], bins=25)

# Add title
plt.title('Number of participants', fontweight='bold',fontsize=FS)
#plt.style.use('seaborn')

# Add data labels
ax.set_xlabel('Number of participants',fontsize=FS-2)
ax.set_ylabel('Number of studies',fontsize=FS-2)


# Set the ranges to plot of x
ax.set_xlim([0, 220])
ax.set_ylim([0, 35])


ax.tick_params(axis='both', which='major', labelsize=16)
ax.xaxis.set_minor_locator(MultipleLocator(5))
ax.tick_params(which='minor', length=3, color='k', direction='out')
ax.tick_params(which='major', length=5, color='k')
plt.show()
```
